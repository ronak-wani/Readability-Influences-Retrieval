```python
import json
import textstat
import pandas as pd
import sys

with open('/content/answers (1).json') as f:
        data = json.load(f)

metrics = []
for topic, source_texts in data.items():
    for source_text_type, answers in source_texts.items():
        for answer_type, response_text in answers.items():


            fk_grade = textstat.flesch_kincaid_grade(response_text)
            ari_score = textstat.automated_readability_index(response_text)
            question_type = "Adv-Q"
            if(answer_type == "Adv-A"):
                question_type = "Adv-Q"
            elif(answer_type == "Int-A"):
                question_type = "Int-Q"
            else:
                question_type = "Ele-Q"

            metrics.append({
                'Source_Text_Type': source_text_type,
                'Question_Type': question_type,
                'FK_Grade': fk_grade,
                'ARI_Score': ari_score,
            })


df = pd.DataFrame(metrics)


print("\n\nAverage Flesch-Kincaid Grade Level Matrix\n\n")
fk_matrix = df.pivot_table(
    values='FK_Grade',
    index='Source_Text_Type',
    columns='Question_Type',
    aggfunc='mean'
)
print(fk_matrix)

print("\n\n\n\n\nAverage Automated Readability Index Matrix\n\n")
ari_matrix = df.pivot_table(
    values='ARI_Score',
    index='Source_Text_Type',
    columns='Question_Type',
    aggfunc='mean'
)
print(ari_matrix)
```
